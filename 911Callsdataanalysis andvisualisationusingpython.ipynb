```python
df['timeStamp'] = pd.to_datetime(df['timeStamp']) #converting to datetime 
```

```python
df.info()
```

```python
time=df['timeStamp'].iloc[0]
time.hour
```

```python
time.dayofweek
```

```python
df['Hour']=df['timeStamp'].apply(lambda time: time.hour)
```

```python
df['Month']=df['timeStamp'].apply(lambda time: time.month)
```

```python
df['DayofWeek']=df['timeStamp'].apply(lambda time: time.dayofweek)
```

```python
df.head()
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'} #mappingtodayofweek
```

```python
df['DayofWeek'] = df['DayofWeek'].map(dmap) #mapping based on dictionary
```

```python
df
```

```python
sns.countplot(x= 'DayofWeek', hue = 'Reason', data = df) #Visualisation of days of week based on Reason
# To relocate the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
